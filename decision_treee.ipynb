```python
from sklearn.datasets import load_iris 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
iris_dataset=load_iris()
```

```python
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(iris_dataset.data,
                                               iris_dataset.target,
                                               stratify=iris_dataset.target,
                                               random_state=22,
                                               shuffle=True)
```

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train_std=scaler.transform(X_train)
y_train_std=scaler.transform(X_test)
```

```python
print(np.bincount(iris_dataset.target))
print(np.bincount(y_train))
print(np.bincount(y_test))
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
model=tree.DecisionTreeClassifier()
model=model.fit(X_train,y_train)
```

```python
tree.plot_tree(model)
```

using entropy as spliting criteria

```python
model1=tree.DecisionTreeClassifier(criterion='entropy')
model1=model1.fit(X_train,y_train)
tree.plot_tree(model1)
```

```python
tree.plot_tree(model)
```

```python
print(accuracy_score(y_test,model.predict(X_test)))
print(accuracy_score(y_test,model1.predict(X_test)))
```

```python
dot_data=export_graphviz(model,class_names=['Stecosa','versicolor','verginicia'])
graph=graph_from_dot_data(dot_data)
graph.write_png('tree.png')
```
